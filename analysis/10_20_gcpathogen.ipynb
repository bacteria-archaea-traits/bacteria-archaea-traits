```python
import requests
import json
URL = "https://nmdc.cn/gcpathogenapi/species/getallmetatable"
```

```python
def constructor_payload_params(start_page, page_size ):
    payload = json.loads('{"otherType":"taxa","page":1,\
                    "high":"","searchStrainName":"","searchAssembly":"",\
                    "searchAssemblyLevel":"","searchAssemblyMethod":"",\
                    "searchContigsEnd":"","searchContigsStart":"",\
                    "searchCountry":"","searchDisease":"",\
                    "searchHost":"","searchGene":"","searchIsoSource":"",\
                    "searchLevel":"","searchMLST":"","searchPathogenName":"",\
                    "searchSequencing":"","searchSubmitter":"","searchTaxonid":"",\
                    "searchAroName":"","searchVFName":"","size":1,"type":"bacteria"}')
    payload["page"] = start_page
    payload["size"] = page_size
    return payload

def get_paginate_gcpathogen(start_page, page_size):
    
    payload = constructor_payload_params(start_page, page_size)
    headers = {'Content-Type': 'application/json'}
    response = requests.post(URL, headers=headers, json = payload)
    if response.status_code > 200:
        return None
    json_response = response.json().get("data", None)
    if json_response is None : return None
    return ( 
        int(json_response.get("totalPage")), 
        int(json_response.get("curPage")),
        int(json_response.get("start")),
        int(json_response.get("end")),
        list(json_response.get("list")) )

def get_all_pages(start_page, page_size):
    data = []
    while True:
        result = get_paginate_gcpathogen(start_page, page_size)
        if result is None: 
            break 
        totalPage, curPage, start, end, page_data  = result
        print(f'Processing page {curPage} start: {start} end: {end} total page: {totalPage}')
        data.extend(page_data)
        start_page = curPage + 1
    return data
```

```python
species_data = get_all_pages(start_page= 1, page_size= 20)
```

```python
import pandas as pd
pd.DataFrame(species_data).to_csv("../output/gcpathogen_pathogens.csv")
```
